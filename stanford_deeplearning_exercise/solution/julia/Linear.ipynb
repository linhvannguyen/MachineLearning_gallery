```julia
function read_float_vector (fname::String)
    f = open(fname)
    lines = readlines(f)
    floats = [x |> strip |> float for x in lines]
    convert(Array{Float64,1}, floats)
end
```

```julia
alpha = 0.07
```

```julia
x = read_float_vector("../ex2x.dat")
```

```julia
y = read_float_vector("../ex2y.dat")
```

```julia
using Color
using PyPlot
```

```julia
scatter(x, y)
```

```julia
x_1 = Float64[c==1 ? v : 1.0 for v=x,c=1:2] # We index over two things to infer a 2-d array instead of an array of arrays
```

```julia
theta = [1.0, 2.0]
error = x_1 * theta - y
while abs(sum(error)) > 0.001
  theta = theta - alpha * x_1' * error / length(x)
  error = x_1 * theta - y
end
```

```julia
hold(true)
plot(x, x_1 * theta)
scatter(x,y; color="black")
```
